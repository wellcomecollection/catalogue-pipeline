```python
import sys
import os
import logging
from lxml import etree
# Add src to path if not present
src_path = os.path.abspath(os.path.join(os.getcwd(), '..', 'src'))
if src_path not in sys.path:
    sys.path.append(src_path)

# Configure logging
logging.basicConfig(level=logging.INFO)
logging.getLogger("oai_pmh_client").setLevel(logging.INFO)

from datetime import datetime, timezone

# Select which adapter to use - change this to "folio" or "axiell"
ADAPTER = "axiell"  # Options: "axiell", "folio"

if ADAPTER == "axiell":
    from adapters.axiell.runtime import AXIELL_CONFIG as ADAPTER_CONFIG
    from adapters.axiell import config
elif ADAPTER == "folio":
    from adapters.folio.runtime import FOLIO_CONFIG as ADAPTER_CONFIG
    from adapters.folio import config
else:
    raise ValueError(f"Unknown adapter: {ADAPTER}")

# Construct the OAI-PMH client
client = ADAPTER_CONFIG.build_oai_client()
SET = "collect"
RECORD_ID = "110003654"
print(f"Using {ADAPTER.upper()} adapter")
sets = client.list_sets()
for oaiset in sets:
    print(f"set: {oaiset}")
mdformats = [mdformat.prefix for mdformat in client.list_metadata_formats()]
print(mdformats)
for mdformat in mdformats:
    record = client.get_record(identifier=f"{SET}:{RECORD_ID}", metadata_prefix=mdformat)
    print(mdformat)
    print(etree.tostring(record.metadata, pretty_print=True, encoding="unicode"))
```
